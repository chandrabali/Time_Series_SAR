```python
# not considering entries related to  'black border'
from numba import jit, njit
@jit(parallel=True)

def divide_samples_in_equality_1(x,y,size_of_vocab,num_of_samples,num_of_classes, class_ini_num):
  #100 samples per class 
    X=np.zeros((num_of_samples * num_of_classes ,size_of_vocab))
    Y=np.zeros(num_of_samples * num_of_classes)
    counter=0
    n=class_ini_num
    l=0
    i=0
    
    while (n<=(class_ini_num+num_of_classes-1)): #clas index goes until 10 (4 , 9 omitted)
        while (i< x.shape[0] and counter <num_of_samples):
            
            if(int(y[i]) == n):
              
                X[l]=x[i]
                Y[l]=y[i]
                counter +=1
                i+=1
                l+=1
                
            else:
                i+=1
                
        print(n)
        if(counter < num_of_samples):
            print("less than "+str(num_of_samples) + " "+str(counter) )
        n+=1
        counter=0
        i=0

    return X,Y
```

```python
def set_intersection(minuend, subtrahend):
    # shows the places at minuend where common elements appear
    sub=subtrahend.tolist()
    minu= minuend.tolist()
    common=[]
    for i in range(len(sub)):
        for j in range(len(minu)):
            if(sub[i]==minu[j]):
                common.append(j)
        
    return set(common)
```
